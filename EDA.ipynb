## Exploratory Data Analysis

```python
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
# Read data
scores = pd.read_csv("scores.csv")

# Summarize data
print('Shape of data set: ' + str(scores.shape))
scores.head()
```

```python
scores.describe()
```

```python
# Read data
all_fields = pd.read_csv("all_fields.csv")

# Summarize data
print('Shape of data set: ' + str(all_fields.shape))
all_fields.head()
```

```python
all_fields.describe()
```

```python
# Report for scores

prof = ProfileReport(scores)
prof.to_file(output_file='scores_report.html')
```

```python
# Report for all_fields

prof = ProfileReport(all_fields)
all_fields.profile_report(correlations={"cramers": {"calculate": False}}) #it crashes so needed to be removed
prof.to_file(output_file='all_fields_report.html')
```
